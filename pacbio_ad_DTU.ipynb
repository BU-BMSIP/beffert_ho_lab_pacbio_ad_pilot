```python
# read in TPM normalized counts matrix
counts_tpm = pd.read_csv("./count_matrix/FLNC_TPM_matrix.txt", sep='\t')
counts_tpm
```

```python
# get ensembl ids from pacbio isoform ids

test_ensembl = counts_tpm.head(n = 50)

def get_ensembl_id_from_symbol(symbol):
  server = "https://rest.ensembl.org"
  ext = f"/xrefs/symbol/homo_sapiens/{symbol}?external_db=HGNC&object_type=transcript"

  r = requests.get(server + ext, headers={ "Content-Type" : "application/json"})

  if not r.ok:
    return None

  decoded = r.json()
    
  if len(decoded) > 0:
    return decoded[0].get("id", None)
  else:
    return None


#counts_tpm["ensembl"] = counts_tpm["associated_gene"].apply(get_ensembl_id_from_symbol)

#counts_tpm.head()


test_ensembl["ensembl"] = test_ensembl["associated_gene"].apply(get_ensembl_id_from_symbol)
test_ensembl

```
